## Exercise H5.2: Optimization

In order to apply **online ICA** we first define the individual error 

$$ e^{(\alpha)} = \ln(\det(W)) + \sum_{l = i}^{N} \ln \left(f'\left( W \cdot x^{(\alpha)} \right)_i  \right)$$

where we will use the logistic function $f(z) = \frac{1}{1 + e^{-z}}$ as our activation function. In this case $$f'(z) = f(z)(1 - f(z)) \quad , \quad f^{''}(z) = f'(z)(1 - 2 f(z)) \quad \Rightarrow \quad \frac{f^{''}(z)}{f^{'}(z)} = 1 - 2 f(z)$$ 

This error is not really needed for the optimization step but it becomes handy for keeping track of the generalization error at each step.

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))

def sigmoid_p(z):
    return sigmoid(z)*(1 - sigmoid(z))

def error(W, x):
    
    log_det = np.log(np.abs(np.linalg.det(W)))
    log_f_prime = np.log(sigmoid_p(np.dot(W, x)))
    error = log_det + np.sum(log_f_prime, 0)
    
    return error
```

### (a) Standard Gradient

In order to define an optimization step with standard gradient ascend we need the gradient of the individual error which is given by

$$ \frac{\partial e^{(\alpha)}}{\partial W_{i j}} = (W^{-1})_{j i} + \frac{f^{''}(W \cdot x^{(\alpha)})_i}{f'(W \cdot x^{(\alpha)})_i} x^{(\alpha)}_j = (W^{-1})_{j i} + (1 - 2 f(W \cdot x^{(\alpha)}))_i x^{(\alpha)}_j$$

and the optimization step is just given by $\delta W_{i j} = \eta \frac{\partial e^{(\alpha)}}{\partial W_{i j}}$

### (b) Natural Gradient

On the other hand, for natural gradient we have the following optimization rule

$$ \delta W_{i j} = \eta \left[\frac{\partial e^{(\alpha)}}{\partial W} W^T W\right]_{i j} = W_{i j} + (1 - 2 f(W \cdot x^{(\alpha)}))_i (x^{(\alpha)} \cdot W^T \cdot W)_j$$

```python
def optimizer_GA(W, x):
    
    error_grad = np.linalg.inv(W).T + np.tensordot(1 - 2*sigmoid(W.dot(x)), x, axes = 0)
    
    return error_grad

def optimizer_NG(W, x):
    
    error_grad = W  + np.tensordot(1 - 2*sigmoid(W.dot(x)), x.dot(W.T).dot(W), axes = 0)
    
    return error_grad
```

Then we define the training process which takes a generic optimizer and updates the weights accordingly. We use different update rules for each optimizer

```python
def train(x, optimizer, eta0, epochs, verbose = True, decay = 0.9999):
    
    # initialize unimixing matrix randomly
    W = np.random.randn(N, N)
    
    for epoch in range(epochs + 1):
        # initial learning rate
        eta = eta0
        # loop over observations
        for t in range(len(x[0])):
            # updates learning rate
            eta = eta*decay
            
            xa = x[:, np.random.randint(len(x[0]))] # pick random example
            delta_W = optimizer(W, xa) # computes delta W with gradient ascend
            W += eta * delta_W
        # after some epochs we compute the TOTAL error, i.e. the average over all examples
        if verbose and (epoch < 10 or epoch % 10 == 0):
            loss = np.sum(error(W, x))/len(x[0])
            print(f"Epoch: {epoch}, Loss: {round(loss, 4)}")
    
    return W
```

```python
# training
print("Gradient Ascend\n")
W_opt_GA = train(x_centered, optimizer_GA, eta0 = 0.01, epochs = 30)
print("\nNatural Gradient\n")
W_opt_NG = train(x_centered, optimizer_NG, eta0 = 0.01, epochs = 30)
```

## Exercise H5.3: Results

##### a)

Plot and Play

```python
rate = 8192 # the data was recorded at 8192 Hz
```

```python
# original sounds
fig, ax = plt.subplots(2, 1, sharex = True, figsize = (15, 6))

ax[0].plot(sound_1, 'b-', alpha = 0.7)
ax[0].set_ylabel("$s_1$", rotation = "horizontal", size = 'x-large')
ax[0].grid()
ax[1].plot(sound_2, 'b-', alpha = 0.7)
ax[1].set_ylabel("$s_2$", rotation = "horizontal", size = 'x-large')
ax[1].grid()
plt.xlabel("Time", size = 'x-large')
fig.suptitle("Original Sources", size = 'xx-large')
plt.show()

write("original_source_1", rate, sound_1)
write("original_source_2", rate, sound_2)
```

```python
# mixed sources before permutation
fig, ax = plt.subplots(2, 1, sharex = True, figsize = (15, 6))

ax[0].plot(x[0], 'b-', alpha = 0.7)
ax[0].set_ylabel("$x_1$", rotation = "horizontal", size = 'x-large')
ax[0].grid()
ax[1].plot(x[1], 'b-', alpha = 0.7)
ax[1].set_ylabel("$x_2$", rotation = "horizontal", size = 'x-large')
ax[1].grid()
plt.xlabel("Time", size = 'x-large')
fig.suptitle("Mixed sources", size = 'xx-large')
plt.show()

write("mixed_unshuffled_source_1", rate, x[0])
write("mixed_unshuffled_source_2", rate, x[1])
```

```python
# mixed sources after permutation
fig, ax = plt.subplots(2, 1, sharex = True, figsize = (15, 6))

ax[0].plot(x_shuffled[0], 'b-', alpha = 0.7)
ax[0].set_ylabel("$x_1$", rotation = "horizontal", size = 'x-large')
ax[0].grid()
ax[1].plot(x_shuffled[1], 'b-', alpha = 0.7)
ax[1].set_ylabel("$x_2$", rotation = "horizontal", size = 'x-large')
ax[1].grid()
fig.suptitle("Mixed sources after permutation", size = 'xx-large')
plt.show()

write("mixed_shuffled_source_1", rate, x_shuffled[0])
write("mixed_shuffled_source_2", rate, x_shuffled[1])
```

```python
# recovered signal from unshuffled data
s_recovered_GA = W_opt_GA.dot(x) # gradient ascend

fig, ax = plt.subplots(2, 1, sharex = True, figsize = (15, 6))

ax[0].plot(sound_1, 'b-', label = "$s_1$", alpha = 0.7)
ax[0].plot(s_recovered_GA[0], 'r-', label = "$\hat{s}_1$", alpha = 0.7)
ax[0].legend()
ax[0].grid()
ax[1].plot(s_recovered_GA[1], 'r-', label = "$\hat{s}_2$", alpha = 0.7)
ax[1].plot(sound_2, 'b-', label = "$s_2$", alpha = 0.7)
ax[1].grid()
ax[1].legend()

plt.xlabel("Time")
fig.suptitle("Recovered signals with gradient ascend", size = 'xx-large')
plt.show()

write("recovered_source_1_GA", rate, s_recovered_GA[0])
write("recovered_source_2_GA", rate, s_recovered_GA[1])
```

```python
# recovered signal from unshuffled data
s_recovered_NG = W_opt_NG.dot(x) # natural gradient

fig, ax = plt.subplots(2, 1, sharex = True, figsize = (15, 6))

ax[0].plot(sound_1, 'b-', label = "$s_1$", alpha = 0.7)
ax[0].plot(s_recovered_NG[0], 'r-', label = "$\hat{s}_1$", alpha = 0.7)
ax[0].legend()
ax[0].grid()
ax[1].plot(s_recovered_NG[1], 'r-', label = "$\hat{s}_2$", alpha = 0.7)
ax[1].plot(sound_2, 'b-', label = "$s_2$", alpha = 0.7)
ax[1].grid()
ax[1].legend()

plt.xlabel("Time")
fig.suptitle("Recovered signals with natural gradient", size = 'xx-large')
plt.show()

write("recovered_source_1_NG", rate, s_recovered_NG[0])
write("recovered_source_2_NG", rate, s_recovered_NG[1])
```
