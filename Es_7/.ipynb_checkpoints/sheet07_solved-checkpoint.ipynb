## Exercise H7.1: Simulated Annealing

We begin by defining the cost function, which in our case corresponds to the energy of the system, and the Gibbs probabilty which quantifies the probability of changing states.

```python
# Energy function
def E(s, W):
    return - 1/2 * s.dot(W).dot(s)

# Gibbs distribution
def Gibbs(beta, DeltaE):
    return 1/(1 + np.e**(beta * DeltaE))
```

### A) Initialization

```python
# hyperparameters
N = 6
t_max = 300
tau = 1.01

# random initial state in {-1, 1}^6
s0 = np.array([1 if x > 0.5 else -1 for x in np.random.rand(N)])
print(f"Initial state:\n{s0}\n")

# random weigths symmetric and with non diagonal elements
W = 2*np.random.rand(N, N) - 1
W = 1/2*(W + W.T) # symmetric
for i in range(N): W[i, i] = 0 # zero diagonal
print(f"Initial weights:\n{np.round(W, 2)}")
```

### B) Optimization

For the optimization problem we need to change the state following the Gibbs-Boltzmann distribution.

```python
def simulated_annealing(M, beta0, tau = tau, t_max = t_max):
    
    s = s0.copy()
    beta = beta0
    Energy = [E(s, W)]
    Temperature = [1/beta0]
    
    for t in range(t_max):

        for m in range(M):
            node = np.random.randint(N)
            Delta_E = 2 * E(s, W)
            # change s_node with probability given by the Gibbs distribution
            flip_prob = 1 - Gibbs(beta, Delta_E)
            s[node] = np.random.choice([-s[node], s[node]], 1, p = [flip_prob, 1 - flip_prob])

        beta = tau*beta
        Energy.append(E(s, W))
        Temperature.append(1/beta)
        
    return s, Energy, Temperature
```

### C) Visualization

```python
s_1, E_1, T_1 = simulated_annealing(M = 1, beta0 = 0.1)
s_20, E_20, T_20 = simulated_annealing(M = 20, beta0 = 0.1)

fig, (ax1, ax2) = plt.subplots(2, 1, sharex = True, figsize = (16, 8))

ax1.plot(E_1, 'r-', label = 'Energy $M=1$')
ax1.plot(E_20, 'b-', label = 'Energy $M=20$')
ax1.legend()
ax1.set_title("Energy vs iterations")
ax1.grid()

ax2.plot(T_1, 'b-')
ax2.set_title("Temperature vs iterations")
ax2.grid()

plt.show()
```

```python
all_states = all_states = list(map(np.array, list(product([1,-1], repeat = N))))
all_energies = np.array([E(state, W) for state in all_states])
num_states = len(all_states)
```

```python
plt.figure(figsize=(16,6))
plt.bar(range(num_states), all_energies)
plt.grid()
plt.xticks(range(num_states))
plt.title("Energy of all possible configurations")
plt.ylabel("E(s)", rotation = 'horizontal')
plt.xlabel("s")
plt.show()
```

```python
# probability for the network of been in a particular state
def Prob(beta, E):
    Z = np.sum(np.e**(-beta*all_energies))
    return np.e**(-beta*E)/Z
```

```python
betas = [0.02, 0.2, 2]
prob_1 = np.array([Prob(betas[0], energy) for energy in all_energies])
prob_2 = np.array([Prob(betas[1], energy) for energy in all_energies])
prob_3 = np.array([Prob(betas[2], energy) for energy in all_energies])
all_probs = [prob_1, prob_2, prob_3]
```

```python
fig, axes = plt.subplots(3, 1, sharex = True, figsize=(14, 8))

for i in range(3):
    axes[i].bar(range(num_states), all_probs[i])
    axes[i].set_title(f"beta = {betas[i]}")
    

plt.xticks(range(num_states))
plt.suptitle("Probability distribution at different temperatures")
plt.tight_layout()
plt.show()
```

## Exercise H7.2: Mean-Field Annealing

### A) Initialization

```python
# hyperparameters
N = 6
t_max = 300
tau = 1.01
eps = 0.1

# random initial first moments
s0_mean = 2*np.random.rand(N) - 1
print(f"Initial first moments <s0>:\n{np.round(s0_mean,2)}\n")

# the random weigths are taken as same as the previous exercise
print(f"Initial weights:\n{np.round(W, 2)}")
```

### B) Optimization

```python
def mean_field_annealing(beta0, tau = tau, t_max = t_max, eps = eps):
    
    s_mean = s0_mean.copy()
    beta = beta0
    Energy = [E(s_mean, W)]
    Temperature = [1/beta0]
    
    for t in range(t_max):
        
        while True:
            
            e_old = - W.dot(s_mean)
            s_mean = np.tanh(-beta*e_old)
            e_new = - W.dot(s_mean)

            if np.all(abs(e_new - e_old) < eps):
                break
                
        beta = tau*beta
        Energy.append(E(s_mean, W))
        Temperature.append(1/beta)
        
    return s_mean, Energy, Temperature
```

### C) Visualization

```python
s_mean, Energy, Temp = mean_field_annealing(beta0 = 1, tau = 1.5, t_max = 10)

fig, (ax1, ax2) = plt.subplots(2, 1, sharex = True, figsize = (16, 8))

ax1.plot(Energy, 'r-')
ax1.set_title("Energy vs iterations")
ax1.grid()

ax2.plot(Temp, 'b-')
ax2.set_title("Temperature vs iterations")
ax2.grid()

plt.show()
```
